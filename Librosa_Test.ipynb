```python
import librosa as lr
import matplotlib.pyplot as plt
import librosa.display
import soundfile
from IPython.display import Audio
import numpy as np
```

```python
audio_data = "EipA_img/king_war_no_more.wav"
audio_noise = "EipA_img/king_war_no_more_noise.wav"
```

```python
x,sr = lr.load(audio_data)
```

```python
plt.figure(figsize=(10, 5))
#plotting the sampled signal
lr.display.waveplot(x, sr=sr)
```

```python
soundfile.write(audio_noise,np.random.randn(len(x),1)*0.08,sr)
y,sr = lr.load(audio_noise)
print(sr)
```

```python
plt.figure(figsize=(10, 5))
#plotting the sampled signal
lr.display.waveplot(y, sr=sr)
plt.savefig("rauschen.png")
print(len(x),len(y))
print(x)
```

```python
noisy = x+y
plt.figure(figsize=(10, 5))
#plotting the sampled signal
lr.display.waveplot(noisy, sr=sr)
```
